```python
# # SaaS Revenue and Churn Analytics
# # Import Required Libraries

import pandas as pd
import numpy as np
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')
```

```python
# # Load Raw CSV Files

accounts = pd.read_csv("accounts.csv")
churn_events = pd.read_csv("churn_events.csv")
feature_usage = pd.read_csv("feature_usage.csv")
subscriptions = pd.read_csv("subscriptions.csv")
support_tickets = pd.read_csv("support_tickets.csv")
```

```python
# # Initial Data Inspection

def inspect(df, name):
    print(f"\n{name} Shape:", df.shape)
    print(f"\n{name} Missing Values:\n", df.isnull().sum())
    
inspect(accounts, "Accounts")
inspect(churn_events, "Churn Events")
inspect(feature_usage, "Feature Usage")
inspect(subscriptions, "Subscriptions")
inspect(support_tickets, "Support Tickets")
```

```python
# # Accounts Table

accounts.head()
```

```python
# # Accounts Table Cleaning

df = pd.read_csv("accounts.csv")

df.columns = df.columns.str.strip()

df['signup_date'] = pd.to_datetime(df['signup_date'], errors='coerce')
df['seats'] = pd.to_numeric(df['seats'], errors='coerce')
df['is_trial'] = df['is_trial'].astype(str).str.strip().map({'True': True, 'False': False})
df['churn_flag'] = df['churn_flag'].astype(str).str.strip().map({'True': True, 'False': False})

df = df.drop_duplicates(subset='account_id')

df['industry'] = df['industry'].str.strip()
df['country'] = df['country'].str.strip().str.upper()
df['plan_tier'] = df['plan_tier'].str.strip().str.title()

print(df.shape)
print(df.info())
```

```python
# # Subscription Table

subscriptions.head()
```

```python
# # Subscriptions Table Cleaning

df = pd.read_csv("subscriptions.csv")

df.columns = df.columns.str.strip()

df['start_date'] = pd.to_datetime(df['start_date'], errors='coerce')
df['end_date'] = pd.to_datetime(df['end_date'], errors='coerce')

df['seats'] = pd.to_numeric(df['seats'], errors='coerce')
df['mrr_amount'] = pd.to_numeric(df['mrr_amount'], errors='coerce')
df['arr_amount'] = pd.to_numeric(df['arr_amount'], errors='coerce')

bool_cols = ['is_trial', 'upgrade_flag', 'downgrade_flag', 'churn_flag', 'auto_renew_flag']
for col in bool_cols:
    df[col] = df[col].astype(str).str.strip().map({'True': True, 'False': False})

df = df.drop_duplicates(subset='subscription_id')

df.loc[df['end_date'] < df['start_date'], 'end_date'] = pd.NaT

df['arr_amount'] = df['mrr_amount'] * 12

df['plan_tier'] = df['plan_tier'].str.strip().str.title()
df['billing_frequency'].str.strip().str.lower()

print(df.duplicated().sum())
print(df.info())
print(df.describe())
```

```python
# # Feature Usage 

feature_usage.head()
```

```python
# # Feature Usage Table Cleaning

df = pd.read_csv("feature_usage.csv")

df.columns = df.columns.str.strip()

df['usage_date'] = pd.to_datetime(df['usage_date'], errors='coerce')

num_cols = ['usage_count', 'usage_duration_secs', 'error_count']
for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df['is_beta_feature'] = df['is_beta_feature'].astype(str).str.strip().map({'True': True, 'False': False})

df = df.drop_duplicates(subset='usage_id')

df = df[df['usage_count'] >= 0]
df = df[df['usage_duration_secs'] >= 0]
df = df[df['error_count'] >= 0]

df['feature_name'] = df['feature_name'].str.strip().str.lower()

df['avg_usage_time_per_event'] = df['usage_duration_secs'] / df['usage_count']
df['avg_usage_time_per_event'] = df['avg_usage_time_per_event'].replace([float('inf')], 0)

print(df.duplicated().sum())
print(df.info())
print(df.describe())
```

```python
# # Support Tickets

support_tickets.head()
```

```python
# # Support Tickets Table Cleaning

df = pd.read_csv("support_tickets.csv")

df.columns = df.columns.str.strip()

df['submitted_at'] = pd.to_datetime(df['submitted_at'], errors='coerce')
df['closed_at'] = pd.to_datetime(df['closed_at'], errors='coerce')

df['resolution_time_hours'] = pd.to_numeric(df['resolution_time_hours'], errors='coerce')
df['first_response_time_minutes'] = pd.to_numeric(df['first_response_time_minutes'], errors='coerce')
df['satisfaction_score'] = pd.to_numeric(df['satisfaction_score'], errors='coerce')

df['escalation_flag'] = df['escalation_flag'].astype(str).str.strip().map({ 'True': True, 'False': False})

df['priority'] = df['priority'].str.strip().str.lower()

df['satisfaction_score'] = df['satisfaction_score'].fillna(df['satisfaction_score'].mean())

df = df.drop_duplicates(subset='ticket_id')

df.loc[df['closed_at'] < df['submitted_at'], 'closed_at'] = pd.NaT

df = df[df['resolution_time_hours'] >= 0]
df = df[df['first_response_time_minutes'] >= 0]

print(df.duplicated().sum())
print(df.info())
print(df.describe())
```

```python
# # Churn Events

churn_events.head()
```

```python
# # Churn Events Table Cleaning

df = pd.read_csv("churn_events.csv")

df.columns = df.columns.str.strip()

df['churn_date'] = pd.to_datetime(df['churn_date'], errors='coerce')

df['refund_amount_usd'] = pd.to_numeric(df['refund_amount_usd'], errors='coerce')

bool_cols = ['preceding_upgrade_flag', 'preceding_downgrade_flag', 'is_reactivation']
for col in bool_cols:
    df[col] = df[col].astype(str).str.strip().map({'True': True, 'False': False})

df['reason_code'] = df['reason_code'].str.strip().str.lower()
df['reason_code'] = df['reason_code'].replace({'unknown': 'other'})

df['feedback_text'] = df['feedback_text'].fillna('no feedback')
df['refund_amount_usd'] = df['refund_amount_usd'].fillna(0)

df = df.drop_duplicates(subset='churn_event_id')

df = df[~((df['preceding_upgrade_flag'] == True) & (df['preceding_downgrade_flag'] == True))]

df['feedback_text'] = df['feedback_text'].str.lower().str.strip()

print(df.duplicated().sum())
print(df.info())
print(df.describe())
```

```python
# # Export Clean data

accounts.to_csv("accounts_clean1.csv", index=False)
subscriptions.to_csv("subscriptions_clean1.csv", index=False)
feature_usage.to_csv("feature_usage_clean1.csv", index=False)
support_tickets.to_csv("support_tickets_clean1.csv", index=False)
churn_events.to_csv("churn_events_clean1.csv", index=False)
```
